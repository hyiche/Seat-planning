# classroom infomation

```python
class Classroom_info(object):
    def __init__(self,  classroomID, rows, columns,  capacity, courseID, group, department, teacher, dataframe):
        self.rows = rows
        self.group = group
        self.df = dataframe
        self.columns = columns
        self.teacher = teacher
        self.capacity = capacity
        self.courseID = courseID
        self.department = department
        self.classroomID = classroomID

    def compare_two_df(self, df1, df2):
        """比兩dataframe的長度大小，使較小的始終在回傳的第一位置"""
        if len(df1) <= len(df2):
            return df1, df2
        else:
            return df2, df1

    def padding_empty_row(self, df1, df2):
        """在一間教室有兩個班級的情況下，為便於後續將前後兩dataframe安排進表格內，故在兩dataframe長度加總小於capacity時
                    補上空白欄進後面的dataframe中。"""
        df1, df2 = df1.reset_index(drop=True), df2.reset_index(drop=True)
        df1, df2 = self.compare_two_df(df1=df1, df2=df2)
        len1, len2 = len(df1), len(df2)
        if len1 + len2 < self.capacity:
            d = self.capacity - len2 - len1
            empty_dataframe = DataFrame([[''] * 2 for _ in range(d)], columns=['Student_ID', 'Name'])
            df2 = df2.append(empty_dataframe, ignore_index=True)
        else:
            assert len1 + len2 == self.capacity
        return df1, df2

    def add_serial_both_df(self, df1, df2):
        """加上正確的 serial number 給兩dataframe，供後續藉由除法運算得商數、餘數排進表格內。"""
        df1, df2 = self.compare_two_df(df1=df1, df2=df2)
        len1, len2 = len(df1), len(df2)
        serial = concat([DataFrame([i], columns=['Serial_number']) for i in range(1, len1 + 1)], ignore_index=True)
        df1 = concat([serial, df1], axis=1)
        serial = concat([DataFrame([i], columns=['Serial_number']) for i in range(len1 + 1, len1 + len2 + 1)],
                        ignore_index=True)
        df2 = concat([serial, df2], axis=1)
        return df1.values, df2.values

    def modify_df(self):
        """預處理學生名條，使之長度為capacity，並補上正確的serial number使能夠分配到正確的表格欄位。"""
        if isinstance(self.df, list):  # self.df若是list資料型態，表示一間教室有多個課程班級在裡頭考試
            df1, df2 = self.padding_empty_row(df1=self.df[0], df2=self.df[1])
            self.df[0], self.df[1] = self.add_serial_both_df(df1=df1, df2=df2)
        else:  # self.df若是pandas.dataframe資料型態，表示一間教室只有一個課程班級在裡頭考試
            self.df = self.df.reset_index(drop=True)
            if len(self.df) < self.capacity:  # padding empty row and add serial number.
                d = self.capacity - len(self.df)
                empty_dataframe = DataFrame([[''] * 2 for _ in range(d)], columns=['Student_ID', 'Name'])
                self.df = self.df.append(empty_dataframe, ignore_index=True)
                serial = concat([DataFrame([i], columns=['Serial_number']) for i in range(1, len(self.df) + 1)],
                                ignore_index=True)
                self.df = concat([serial, self.df], axis=1)
                self.df = self.df.values
            else:  # just need to add serial number.
                assert len(self.df) == self.capacity
                serial = concat([DataFrame([i], columns=['Serial_number']) for i in range(1, len(self.df) + 1)],
                                ignore_index=True)
                self.df = concat([serial, self.df], axis=1)
                self.df = self.df.values
```

# load data and note

```python
def load_note_and_data():
    window = Tk()
    window.withdraw()
    print('請於視窗中選取 exam_note.xlsx 的檔案路徑')
    note_path = askopenfilename(title='請選取 exam_note.xlsx 的檔案路徑')
    print('請於視窗中選取 exam.xlsm 的檔案路徑')
    data_path = askopenfilename(title='請選取 exam.xlsm 的檔案路徑')
    note0 = read_excel(note_path, sheet_name=0)  # The dataframe record course info mapping to classroom(s).
    note1 = read_excel(note_path, sheet_name=1)  # The dataframe record classroom info mapping to courseID(s).
    data = read_excel(data_path, sheet_name=1)   # The dataframe record information of all students.
    return note0, note1, data
```

# store infomation to class Classroom_info

```python
def store_info_to_class(note0, note1, data):
    """dic_course: is a dictionary datatype to make courseID map to Department、Teacher、Group information(s).
            dic_classroom:  is a dictionary datatype to make classroomID map to Rows、Columns、Capacity"""

    # filter the classroom(s) which do not arrange to exam this time.
    not_use_classroomID_list = list(note1.Classroom_ID[note1.Content_1.isnull()])
    note1_nonna = note1.dropna(subset=['Content_1'])  # filter the classroom(s) which do not arrange to exam.

    # make courseID map to Department、Teacher、Group information(s)
    dic_course = {str(note0.iloc[i].Course_ID): [note0.iloc[i].Department, note0.iloc[i].Teacher, note0.iloc[i].Group]
                  for i in range(len(note0))}

    # make classroomID map to Rows、Columns、Capacity
    dic_classroom = {note1_nonna.iloc[j].Classroom_ID: [note1_nonna.iloc[j].Rows, note1_nonna.iloc[j].Columns,
                     note1_nonna.iloc[j].Capacity] for j in range(len(note1_nonna))}

    # use dic_classroom to store dataframe(students' information) for each classroomID.
    for i in range(len(note0)):
        i_course = note0.iloc[i]
        underline_idx = i_course.Classroom1.index('_')
        df = data[data['Course_ID'] == int(i_course.Course_ID)][["Course_ID", "Student_ID", "Name"]]

        # 以下條件判斷建立於一個課程最多只會拆分安排在兩間教室進行考試，若規則有更改，需改動程式碼。
        if i_course.Classroom2 is nan:  # 情況一: 若一個課程被分在同一間教室考試。
            dic_classroom[i_course.Classroom1[:underline_idx]].append(df)
        else:  # 情況二: 若一個課程被分到兩間教室考試。
            df_split = split(df, [int(i_course.Classroom1[underline_idx + 1:])], axis=0)
            dic_classroom[i_course.Classroom1[:underline_idx]].append(df_split[0])
            dic_classroom[i_course.Classroom2[:underline_idx]].append(df_split[1])

    # Change the information(s) what I need,from the dictionary data type to Classroom_info class by list comprehension.
    classroom_infomation = [Classroom_info(
            classroomID=i,
            rows=dic_classroom[i][0],
            columns=dic_classroom[i][1],
            capacity=dic_classroom[i][2],
            courseID=str(dic_classroom[i][3].Course_ID.iloc[0]),
            dataframe=dic_classroom[i][3][['Student_ID', 'Name']],
            group=dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][2],
            teacher=dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][1],
            department=dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][0]) if len(dic_classroom[i]) <= 4 else
        Classroom_info(
            classroomID=i,
            rows=dic_classroom[i][0],
            columns=dic_classroom[i][1],
            capacity=dic_classroom[i][2],
            courseID=(str(dic_classroom[i][3].Course_ID.iloc[0]), str(dic_classroom[i][4].Course_ID.iloc[0])),
            dataframe=[dic_classroom[i][3][['Student_ID', 'Name']], dic_classroom[i][4][['Student_ID', 'Name']]],
            group=(dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][2], dic_course[str(dic_classroom[i][4]\
                                                                                         .Course_ID.iloc[0])][2]),
            teacher=(dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][1], dic_course[str(dic_classroom[i][4]\
                                                                                           .Course_ID.iloc[0])][1]),
            department=(dic_course[str(dic_classroom[i][3].Course_ID.iloc[0])][0], dic_course[str(dic_classroom[i][4]\
                                                                                              .Course_ID.iloc[0])][0]))
        for i in list(dic_classroom.keys())]
    return classroom_infomation, not_use_classroomID_list
```

# output to excel

```python
def output_to_Excel(classroom_info, filter_record, time_info):
    # Define the file name of output excel file.
    print('請於視窗中選取輸出檔案的儲存路徑，並在檔案名稱欄位中給定檔名(不須給予副檔名)')
    savepath = asksaveasfilename(title='請選取輸出檔案的儲存路徑，並在檔案名稱欄位中給定檔名(不須給予副檔名)')
    workbook = Workbook(savepath + '.xlsx')
    f = XlsxWriter_Format_Manual_Definition(workbook=workbook)

    for i_classroom in classroom_info:
        # 預處理學生名條，使之長度為capacity，並補上正確的serial number使能夠正確分配到表格欄位。
        i_classroom.modify_df()
        assert i_classroom.rows * i_classroom.columns == i_classroom.capacity
        worksheet = workbook.add_worksheet(i_classroom.classroomID)  # add a worksheet named 'classroomID'.
        # We can only write simple types to merged ranges so we write a blank string.
        worksheet.merge_range(0, 0, 0, i_classroom.columns, "", cell_format=f.format1)
        worksheet.merge_range(1, 0, 1, i_classroom.columns, "", cell_format=f.format2)
        worksheet.set_default_row(height=26.3)
        worksheet.set_row(row=0, height=48, cell_format=f.format1)
        worksheet.set_row(row=1, height=48, cell_format=f.format2)
        worksheet.set_row(row=3, height=15.8, cell_format=f.format8)
        worksheet.set_column('A:A', 1.80, cell_format=f.format8)
        worksheet.set_column(1, i_classroom.columns, 15.20)
        if isinstance(i_classroom.df, list):
            worksheet.write_rich_string('A1', time_info + ' 微積分會考座位表\n', f.red,
                                        '*請確認自己的學號、姓名、應考教室、座位後，簽名確認，謝謝。', f.format1)
            worksheet.write_rich_string('A2', f.black, '綜合大樓' + i_classroom.classroomID, f.blue, ' ' +
                                        i_classroom.department[0] + ' ' + i_classroom.teacher[0] + ' 課程代碼：' +
                                        i_classroom.courseID[0] + '(' + i_classroom.group[0] + ')\n', f.green,
                                        ' '*13 + i_classroom.department[1] + ' ' + i_classroom.teacher[1] +
                                        ' 課程代碼：' + i_classroom.courseID[1] + '(' + i_classroom.group[1] + ')')
            worksheet.write('E3', '講台', f.podium_format)
            # set index of columns in sheet.
            for i in range(i_classroom.columns):
                worksheet.write(3, i + 1, i + 1, f.format8)
            # set index of rows in sheet.
            for i in range(i_classroom.rows):
                worksheet.write(5 + 3 * i, 0, i + 1, f.format8)
            # To write the form and grid.
            for i in i_classroom.df[0]:
                row_idx, col_idx = (i[0] - 1) // i_classroom.columns, (i[0] - 1) % i_classroom.columns + 1
                worksheet.write(4 + 3 * row_idx, col_idx, i[1], f.format3_blue)   # To write studentID cell.
                worksheet.write(6 + 3 * row_idx, col_idx, '', f.format4)          # To write signature cell.
                worksheet.write(5 + 3 * row_idx, col_idx, i[2], f.format6)        # To write student name cell.
            # To write the form and grid.
            for i in i_classroom.df[1]:
                row_idx, col_idx = (i[0] - 1) // i_classroom.columns, (i[0] - 1) % i_classroom.columns + 1
                worksheet.write(4 + 3 * row_idx, col_idx, i[1], f.format3_green)  # To write studentID cell.
                worksheet.write(6 + 3 * row_idx, col_idx, '', f.format4)          # To write signature cell.
                worksheet.write(5 + 3 * row_idx, col_idx, i[2], f.format7)        # To write student name cell.
        else:
            worksheet.write_rich_string('A1', time_info + ' 微積分會考座位表\n', f.red,
                                        '*請確認自己的學號、姓名、應考教室、座位後，簽名確認，謝謝。', f.format1)
            worksheet.write('A2', '綜合大樓' + i_classroom.classroomID + ' ' + i_classroom.department + ' ' +
                            i_classroom.teacher + ' 課程代碼：' + i_classroom.courseID + '(' + i_classroom.group + ')',
                            f.format2)
            worksheet.write('E3', '講台', f.podium_format)
            # set index of columns in sheet.
            for i in range(i_classroom.columns):
                worksheet.write(3, i + 1, i + 1, f.format8)
            # set index of rows in sheet.
            for i in range(i_classroom.rows):
                worksheet.write(5 + 3 * i, 0, i + 1, f.format8)
            # To write the form and grid.
            for i in i_classroom.df:
                row_idx, col_idx = (i[0] - 1) // i_classroom.columns, (i[0] - 1) % i_classroom.columns + 1
                worksheet.write(4 + 3 * row_idx, col_idx, i[1], f.format3)  # To write studentID cell.
                worksheet.write(6 + 3 * row_idx, col_idx, '', f.format4)    # To write signature cell.
                worksheet.write(5 + 3 * row_idx, col_idx, i[2], f.format5)  # To write student name cell.
        print('sheet: ' + i_classroom.classroomID + '，已完成')
    workbook.close()
    if len(filter_record) > 0:  # 可能有空教室或沒有空教室
        print("此次規劃有 %d 間空教室: " % (len(filter_record)) + '、'.join(filter_record))
    print('檔案輸出路徑為: ' + savepath + '.xlsx')
```

# main function

```python
if __name__ == "__main__":
    input('2019.12.01版本，請按 Enter 鍵繼續')
    time_info = input('請輸入此次規劃進行考試的時間日期 Ex: 2019.11.28 16:10-17:00 --->')
    note0, note1, data = load_note_and_data()
    classroom_info, filter_record = store_info_to_class(note0=note0, note1=note1, data=data)
    output_to_Excel(classroom_info=classroom_info, filter_record=filter_record, time_info=time_info)
    input('請按 Enter 鍵結束程式')
```
